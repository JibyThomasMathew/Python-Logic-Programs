```python
# 1. Reverse a string
def reverse_string(s):
    return s[::-1]

# 2. Check for Palindrome
def is_palindrome(s):
    return s == s[::-1]

# 3. Fibonacci Series
def fibonacci(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

# 4. Factorial of a Number
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

# 5. Prime Number Check
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# 6. Count vowels and consonants
def count_vowels_consonants(s):
    vowels = "aeiouAEIOU"
    v = c = 0
    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1
    return v, c

# 7. Sort an array
def sort_array(arr):
    srt_arr = sorted(arr)
    return srt_arr

# 8. Merge two arrays
def merge_arrays(a, b):
    return a + b

# 9. Find the largest element in an array
def largest_element(arr):
    return max(arr)

# 10. Remove duplicates from an array
def remove_duplicates(arr):
    return list(set(arr))

# 11. Check if a number is Armstrong
def is_armstrong(n):
    digits = list(map(int, str(n)))
    power = len(digits)
    return sum(d**power for d in digits) == n

# 12. Reverse a number
def reverse_number(n):
    return int(str(n)[::-1])

# 13. Calculate GCD of two numbers
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# 14. Check for anagram
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)

# 15. Count number of digits
def count_digits(n):
    return len(str(abs(n)))

# 16. Print prime numbers in a range
def primes_in_range(start, end):
    return [i for i in range(start, end + 1) if is_prime(i)]

# 17. Find second largest element
def second_largest(arr):
    unique = list(set(arr))
    unique.sort()
    return unique[-2]

# 18. Swap two numbers
def swap(a, b):
    return b, a

# 19. Pascal's Triangle
def pascal_triangle(n):
    triangle = []
    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i-1][j-1] + triangle[i-1][j]
        triangle.append(row)
    return triangle

# 20. Find missing number in an array
def missing_number(arr, n):
    return n * (n + 1) // 2 - sum(arr)

# 21. Convert decimal to binary
def decimal_to_binary(n):
    return bin(n)[2:]

# 22. Check for perfect number
def is_perfect(n):
    return sum(i for i in range(1, n) if n % i == 0) == n

# 23. Simple calculator
def calculator(a, b, op):
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/': return a / b

# 24. Sum of digits
def sum_of_digits(n):
    return sum(map(int, str(abs(n))))

# 25. Length of string
def string_length(s):
    return len(s)

# 26. Check if string is empty
def is_empty(s):
    return len(s) == 0

# 27. Count occurrences of a character
def char_count(s, ch):
    return s.count(ch)

# 28. First non-repeated character
def first_non_repeated(s):
    for ch in s:
        if s.count(ch) == 1:
            return ch
    return None

# 29. Remove whitespaces
def remove_whitespaces(s):
    return s.replace(" ", "")

# 30. Common elements in two arrays
def common_elements(a, b):
    return list(set(a) & set(b))

# 31. Fibonacci using recursion
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# 32. Generate random number
import random
def random_number(start, end):
    return random.randint(start, end)

# 33. Check leap year
def is_leap_year(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

# 34. Sum of first N natural numbers
def sum_natural(n):
    return n * (n + 1) // 2

# 35. Simple login system
def login(username, password):
    return username == "admin" and password == "1234"

# 36. Check if string contains another string
def contains_string(s, sub):
    return sub in s

# 37. Maximum occurring character
def max_occurring_char(s):
    return max(set(s), key=s.count)

# 38. Selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# 39. Bubble sort
def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```
