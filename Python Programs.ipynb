### Prime Number Check

```python
n = int(input("Enter number: "))
flag = True
if n <= 1:
    flag = False
for i in range(2, int(n**0.5)+1):
    if n % i == 0:
        flag = False
print("Prime" if flag else "Not Prime")
```

### Count number and consonants

```python
s = input("Enter string: ")
vowels = "aeiouAEIOU"
v = c = 0
for ch in s:
    if ch.isalpha():
        if ch in vowels:
            v += 1
        else:
            c += 1
print("Vowels:", v, "Consonants:", c)
```

### Armstrong Number

```python
n = int(input("Enter number: "))
temp = n
sum = 0
power = len(str(n))
while temp > 0:
    d = temp % 10
    sum += d ** power
    temp //= 10
print("Armstrong" if sum == n else "Not Armstrong")
```

### Pascals Triangle

```python
n = int(input())
for i in range(n):
    num = 1
    for j in range(i+1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```

### Selection sort

```python
arr = list(map(int, input().split()))
for i in range(len(arr)):
    min_idx = i
    for j in range(i+1, len(arr)):
        if arr[j] < arr[min_idx]:
            min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
print(arr)
```

### Bubble sort

```python
arr = list(map(int, input().split()))
for i in range(len(arr)):
    for j in range(len(arr)-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]
print(arr)
```
